## VIsualizing highest cancelled reservations for Resort and City hotels

```python
# Checking for Not canceled and canceled reservations in Resort Hotel
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
# Summary: 72% of reservations in Resort hotel wasn't canceled but 28% was canceled
```

```python
# Checking for Not canceled and canceled reservations in City Hotel
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

Summary: 58% of reservations in City hotel weren't canceled but 42% were canceled

```python
# Calculating the mean of Average daily rate (ADR) in respect to reservation status date for both resort and city hotel.

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()

print(resort_hotel)
print(city_hotel)
```

```python
# plotting the Average Daily Rate of City and Resort Hotel

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate of City and Resort Hotel Over the Years', fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

This chart compares the Average Daily Rate (ADR) of Resort Hotels and City Hotels from 2014 to 2017. Resort Hotels typically exhibit higher ADRs than City Hotels during peak periods, such as holidays or vacation seasons, with noticeable variability marked by sharp peaks and dips due to seasonal tourism demand. In contrast, City Hotels maintain a more stable ADR, with minor seasonal fluctuations, likely catering to business travel

```python
# Checking monthly reservation status and months with highest cancellation

df['month'] = df['reservation_status_date'].dt.month
# Plotting
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'dark')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('Months')
plt.ylabel('Number of reservations')
plt.legend(['Not canceled', 'canceled'])
plt.show()
# The chart below shows that when there is less cancellations, the reservations are maximum and when the reseravation are less the cancellations are maximum 
```

This chart shows reservation trends per month, with **not canceled bookings** consistently higher than **canceled bookings**. **July and August** have the highest reservations, reflecting peak travel seasons, but also higher cancellations. **December to February** see fewer reservations overall, aligning with the off-peak season. Higher cancellations during busy months may result from overbooking or travel plan changes.

```python
# Plotting the Average Daily Rate for each month
plt.figure(figsize = (15,8))
plt.title('Average Daily Rate Per Month', fontsize = 30)
# Filtering the DataFrame (df) to include only rows where the value in the 'is_canceled' column is equal to 1.
sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

This chart shows the Average Daily Rate (ADR) per month, with January and July having the highest rates due to peak holiday seasons, while ADRs drop from August to December, reflecting off-peak demand.
